```python
#Importing Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Importing Dataset and Adding Column Names
column_names=['Review','Sentiment','Un2','Un3','Un4','Un5','Un6','Un7','Un8','Un9','Un10','Un11','Un12','Un13','Un14']
df=pd.read_csv('Evaluation-dataset.csv',names=column_names)
df
```

```python
#Feature Names
df.columns
```

```python
# Dimention of Data
df.shape
```

```python
#Finding any null values if present
print(df.isnull().sum())
```

```python
# Counting the Rows of target based on value
df['Sentiment'].value_counts()
```

```python
#Plotting Sentiment feature based on count
plt.figure(figsize=(20,8))
df["Sentiment"].value_counts().plot(kind="bar")
plt.show()
```

```python
#Dropping columns exept first two as the datapointsare mostly having null values
df= df.drop(['Un2','Un3','Un4','Un5','Un6','Un7','Un8','Un9','Un10','Un11','Un12','Un13','Un14'], axis=1)
```

```python
df
```

```python
#Dropping null values
df= df.dropna()
```

```python
df
```

```python
#Renaming datapoints in Sentiment column  as "Positive" if it contains "Positive" in substring
df.loc[df['Sentiment'].str.contains('positive'), 'Sentiment'] = 'Positive'
```

```python
#Renaming datapoints in Sentiment column  as "Negative" if it contains "negative" in substring
df.loc[df['Sentiment'].str.contains('negative'), 'Sentiment'] = 'Negative'
```

```python
#Plotting Sentiment feature based on count after renaming
plt.figure(figsize=(20,8))
df["Sentiment"].value_counts().plot(kind="bar")
plt.show()
```

```python
#Dropping rows in sentiment column if other than positive and negative as they are in few numbers
df = df.drop(df[(df.Sentiment !="Negative") & (df.Sentiment != "Positive")].index)
```

```python
#Plotting Sentiment column after filtering
plt.figure(figsize=(20,8))
df["Sentiment"].value_counts().plot(kind="bar")
plt.show()
```

```python
#Here we can confirm the shape after preprocessing
df.shape
```

```python
df
```

```python
#Encoding string to integer in Sentiment Column
df["Sentiment"]=df["Sentiment"].map({"Positive":1,"Negative":0})
df
```

```python
#Checking value counts
df["Sentiment"].value_counts()
```

```python
# Define X and y data set from "Evaluation  Dataset"
X=df.Review
y=df.Sentiment
print(X.shape)
print(y.shape)
```

```python
# Applying Test Train Split
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y,random_state=7)
```

```python
from collections import Counter
from sklearn.datasets import make_classification
```

# Vectorize text Dataset Using CountVectorizer

```python
# Call the vectorizer
from sklearn.feature_extraction.text import CountVectorizer
toNumeric = CountVectorizer()
```

```python
# Create vocabulary and create document term matrix
toNumeric.fit(X_train)                       ##Create Vocabulary
X_train_dtm = toNumeric.transform(X_train)
```

```python
X_train_dtm
```

```python
#Transform test data using fitted vocabulary into DTM
X_test_dtm = toNumeric.transform(X_test)
X_test_dtm
```

# Build And Evaluate Models

```python
#Naive Bayes 
from sklearn.naive_bayes import MultinomialNB
nb = MultinomialNB()
```

```python
nb.fit(X_train_dtm,y_train)
```

```python
y_pred_class = nb.predict(X_test_dtm)
```

```python
# Accuracy Calculation
from sklearn import metrics
metrics.accuracy_score(y_test,y_pred_class)
```

```python
#Building Confusion Matrix
metrics.confusion_matrix(y_test, y_pred_class)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_class))
```

```python
# Predict probablity for test data instead for labels directly
y_pred_prob = nb.predict_proba(X_train_dtm)[:,1]
y_pred_prob
```

```python
#Accuracy on Test data
nb.score(X_test_dtm,y_test)
```

# Compare other models


```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
lr = LogisticRegression()
```

```python
lr.fit(X_train_dtm,y_train)
```

```python
y_pred_class = lr.predict(X_test_dtm)
```

```python
#Calculate predicted probablities for X_test_dtm
y_pred_prob = lr.predict_proba(X_test_dtm)[:,1]
y_pred_prob
```

```python
metrics.accuracy_score(y_test,y_pred_class)
```

```python
metrics.roc_auc_score(y_test,y_pred_prob)
```

# Fine Tune Vectorizer

```python
# Default parameters for CountVectorizer
toNumeric = CountVectorizer()
toNumeric
```

```python
# Remove English stop words
toNumeric = CountVectorizer(stop_words='english')
toNumeric.fit(X_train)
len(toNumeric.get_feature_names_out())
```

```python
# Include 1 and 2-grams
toNumeric = CountVectorizer(ngram_range=(1, 2))
toNumeric.fit(X_train)
len(toNumeric.get_feature_names_out())
```

```python
# Ignore terms that appear in more than 75% of the documents
toNumeric = CountVectorizer(max_df=0.75)
toNumeric.fit(X_train)
len(toNumeric.get_feature_names_out())
```

```python
# Keep terms that appear in at least 5 documents
toNumeric = CountVectorizer(min_df=2)
toNumeric.fit(X_train)
len(toNumeric.get_feature_names_out())
```
